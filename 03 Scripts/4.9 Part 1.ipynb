## Data Wrangling

```python
#Renaming Columns
customers.rename(columns={'First Name': 'first_name', 'Surnam' : 'surname', 'Gender':'gender', 'STATE': 'state', 'Age':'age','n_dependants':'no_of_dependants','fam_status':'matrial_status'}, inplace= True)
```

```python
#Checking for mixed-type data
for col in customers.columns.tolist():
    weird= (customers[[col]].applymap(type)!= customers[[col]].iloc[0].apply(type)).any(axis =1)
    if len(customers[weird])>0 :
        print(col)
```

```python
#Convert first_name data type to string
customers['first_name'] = customers['first_name'].astype('str')
```

```python
customers.dtypes
```

```python
#Re-checking for mixed-type data
for col in customers.columns.tolist():
    weird= (customers[[col]].applymap(type)!= customers[[col]].iloc[0].apply(type)).any(axis =1)
    if len(customers[weird])>0 :
        print(col)
```

## Reducing data types to save RAM before merge

```python
customers['age'] = customers['age'].astype('int8')
```

```python
customers['no_of_dependants'] = customers['no_of_dependants'].astype('int8')
```

```python
customers['income'] = customers['income'].astype('int32')
```

```python
customers['user_id'] = customers['user_id'].astype('str')
```

```python
customers.dtypes
```

## Merging customers dataframe with ords_prods_merge

```python
#Importing combined data set
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged_3.pkl'))
```

```python
ords_prods_merge.shape
```

```python
ords_prods_merge['user_id'] = ords_prods_merge['user_id'].astype('str')
```

```python
ords_prods_merge.dtypes
```

```python
ords_prods_custs = ords_prods_merge.merge(customers, on = 'user_id')
```

```python
ords_prods_custs.head()
```

```python
ords_prods_custs.shape
```

```python
ords_prods_custs = pd.merge(ords_prods_merge,customers, on = 'user_id', how = 'outer')
```
